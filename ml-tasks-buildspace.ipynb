**For songs with duration duration_ms value less than 300000 ms (300 s), find Top 10 “fastest” songs( songs with highest values for tempo column ) and display values for song_title, artistcolumns.**

```python
import pandas as pd
df=pd.read_csv("/kaggle/input/spotifyclassification/data.csv")
flt_songs = df[df['duration_ms'] < 300000]
top_10_fastest_songs = flt_songs.sort_values(by='tempo', ascending=False).head(10)
top_10_fastest_songs[['song_title', 'artist','tempo']]
```

**Create a dataframe of the songs that rank among top 10 for the instrumentalness metric. This dataframe should only contain the instrumentalness, song title and artist columns.**

**Create a dataframe of the average tempo for each of the top 10 most occurring artists.**

```python
import pandas as pd
df=pd.read_csv("/kaggle/input/spotifyclassification/data.csv")
artist_tempo_means = df.groupby('artist')['tempo'].mean().reset_index()
top_10_artists =artist_tempo_means.sort_values(by='tempo', ascending=False).head(10)
top_10_artists
```

```python
import pandas as pd
df=pd.read_csv("/kaggle/input/spotifyclassification/data.csv")
artist_tempo_means = df.groupby('artist')['tempo'].mean().reset_index()
top_10_artists =artist_tempo_means.sort_values(by='tempo', ascending=False).head(10)
top_10_artists.reset_index(drop=True, inplace=True)
top_10_artists
```
